```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('customer_churn.csv')
df.head()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
#Ages are between 18 to 92
```

```python
labels=['0-20','21-40','41-60','Above 61']
bins=[0, 20, 40, 60, 100]
df['Age_bins']=pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
df.head()
```

```python
y=df['Age_bins'].value_counts()
y
```

```python
bars=plt.bar(x=labels,height=y)
plt.title('Age Bins Distribution ')
plt.xlabel('Age Bins')
plt.ylabel('Counts (Age Bins)')
plt.bar_label(bars)
```
