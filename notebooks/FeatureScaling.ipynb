# Feature Scaling in Machine Learning

Feature Scaling is the process of transforming numerical features into a similar scale.

Why do we need Feature Scaling?

- Some features may have very large values.
- Some features may have very small values.
- Distance-based algorithms are heavily affected by feature magnitude.

Algorithms that require scaling:
- KNN
- K-Means
- SVM
- PCA
- Gradient Descent based models
- Neural Networks

Algorithms that generally do not require scaling:
- Decision Tree
- Random Forest
- XGBoost

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# Loading Customer Churn Dataset

We will use a customer churn dataset to demonstrate Feature Scaling techniques.

The dataset contains:
- Customer demographics
- Account information
- Product usage details
- Churn indicator (Exited)

Our goal is to scale numerical features before model training.

```python
df=pd.read_csv('customer_churn.csv')
df.tail()
```

# Understanding Dataset Structure

The info() function helps us understand:

- Total number of rows
- Total number of columns
- Data types
- Missing values
- Memory usage

Feature Scaling can only be applied to numerical columns.

```python
df.info()
```

# Selecting Numerical Features

Feature Scaling can only be applied to numerical columns.

We use:

select_dtypes(include=['number'])

to extract only numeric variables from the dataset.

```python
#Selecting only numeric columns to scale
numerical_df=df.select_dtypes(include=['number'])
numerical_df.info()
```

```python
numerical_df.head(2)
```

# Removing Identifier Columns

RowNumber and CustomerId are unique identifiers.

These columns do not provide predictive information.

Keeping identifiers may:
- introduce noise
- distort scaling
- negatively impact model performance

Therefore we remove them before scaling.

```python
numerical_df.drop(['RowNumber','CustomerId',], inplace=True, axis=1)
numerical_df.head()
```

# Normalization using Min-Max Scaling

Min-Max Scaling transforms values into a fixed range.

Formula:

X_scaled = (X - X_min) / (X_max - X_min)

Output Range:
- 0 to 1

Advantages:
- Preserves original distribution shape
- Easy interpretation

Commonly used in:
- Neural Networks
- Deep Learning
- Image Processing

```python
mm_scaler=MinMaxScaler()
scaled_arr=mm_scaler.fit_transform(numerical_df)
```

# Viewing Normalized Values

After applying Min-Max Scaling:

- Smallest value becomes 0
- Largest value becomes 1
- All other values fall between 0 and 1

This helps algorithms treat features on a similar scale.

```python
scaled_arr[0:3]
```

# Standardization using Standard Scaler

Standardization transforms data to have:

- Mean = 0
- Standard Deviation = 1

Formula:

Z = (X - Mean) / Standard Deviation

Advantages:
- Handles outliers better than Min-Max Scaling
- Most commonly used scaling technique

Commonly used in:
- Logistic Regression
- SVM
- PCA
- Linear Regression
- Gradient Descent based algorithms

```python
ss=StandardScaler()
ss_arr=ss.fit_transform(numerical_df)
```

# Viewing Standardized Values

After Standardization:

- Positive values are above the mean
- Negative values are below the mean
- Values near 0 are close to the mean

This creates a standard distribution that helps many machine learning algorithms converge faster.

```python
ss_arr[0:5]
```

# Interview Summary

## Min-Max Scaling

Range:
0 to 1

Formula:
(X - Min) / (Max - Min)

Use Cases:
- Neural Networks
- Deep Learning
- Image Data

Advantages:
- Easy interpretation
- Preserves distribution

Disadvantages:
- Sensitive to outliers

---

## Standardization

Mean:
0

Standard Deviation:
1

Formula:
(X - Mean) / Standard Deviation

Use Cases:
- Logistic Regression
- SVM
- PCA
- Linear Regression

Advantages:
- Less affected by outliers
- Most commonly used

Disadvantages:
- Does not bound values between 0 and 1
