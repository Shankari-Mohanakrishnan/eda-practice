```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df=pd.read_csv('customer_churn.csv')
df.tail()
```

```python
df.info()
```

```python
#Selecting only numeric columns to scale
numerical_df=df.select_dtypes(include=['number'])
numerical_df.info()
```

```python
numerical_df.head(2)
```

```python
numerical_df.drop(['RowNumber','CustomerId',], inplace=True, axis=1)
numerical_df.head()
```

```python
#Normalization - MinMax Scaler
```

```python
mm_scaler=MinMaxScaler()
scaled_arr=mm_scaler.fit_transform(numerical_df)
```

```python
scaled_arr[0:3]
```

```python
#Standardization - Standard Scaler
```

```python
ss=StandardScaler()
ss_arr=ss.fit_transform(numerical_df)
```
