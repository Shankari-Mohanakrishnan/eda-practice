## Dropping Columns with Missing Values

The `dropna(axis=1)` method removes entire columns that contain missing values.

Here:
- `axis=1` means columns
- `axis=0` means rows

This method is useful when:
- a column contains too many missing values
- the column is not important for analysis
- imputing values is not practical

Limitation:
- important information may be lost if useful columns are removed.

```python
updated_df=df.dropna(axis=1)
updated_df.info()
```

If we use this method we might be losing valuable information. This method cannot be used if the columns have to many missing values or if they have information required for analysis.

## Dropping Rows with Missing Values

This method is useful when:
- the dataset is large
- missing rows are very few
- removing rows does not impact analysis significantly

```python
updated_df=df.dropna(axis=0)
updated_df.info()
```

## Mean Imputation

Mean imputation replaces missing numerical values using the average value of the column.

Advantages:
- simple and fast
- preserves dataset size

Limitations:
- sensitive to outliers
- may reduce variance

```python
mean_imputation_df=df
mean_imputation_df['Age']=mean_imputation_df['Age'].fillna(df['Age'].mean())
mean_imputation_df.info()
```

## Median Imputation

Median imputation replaces missing values using the median of the column.

This method is preferred when:
- data contains outliers
- distribution is skewed

```python
median_imputation_df=df
median_imputation_df['Age']=median_imputation_df['Age'].fillna(df['Age'].median())
median_imputation_df.info()
```

## Forward Fill

Forward fill propagates the previous valid value forward to replace missing values.

Commonly used in:
- time series data
- sequential datasets
  
## Backward Fill

Backward fill replaces missing values using the next valid observation in the dataset.

```python
df=pd.read_csv('Churn_Modelling.csv')
df.info()
```

```python
df['Age']=df['Age'].bfill()
df.info()
```
