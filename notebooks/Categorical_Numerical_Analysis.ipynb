```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_data=pd.read_csv('customer_churn.csv')
base_data.head()
```

```python
cat_data=base_data[['Geography','Gender','Exited']]
cat_data.head()
```

```python
cat_data['Exited'].value_counts()
```

```python
#Categorical Analysis
```

```python
sns.countplot(data=cat_data, x='Geography')
```

```python
sns.countplot(data=cat_data, x='Gender')
```

```python
sns.countplot(data=cat_data, x='Geography', hue='Exited')
```

```python
sns.countplot(data=cat_data, x='Gender', hue='Exited')
```

```python
sns.histplot(data=base_data, x='Geography', hue='Gender', multiple='dodge')
```

```python
sns.histplot(data=base_data, x='Gender', hue='Geography', multiple='dodge')
```

```python
#Numerical Analysis
```

```python
base_data.head(2)
```

```python
num_data=base_data[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Exited']]
num_data.head()
```

```python
#Correlation - ranges from -1 to +1
```

```python
cor=num_data.corr()
cor
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cor,annot=True)
```

```python
num_data.corr()['Exited'].sort_values(ascending=False).plot(kind='bar')
```

```python
num_data['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
num_data_exited_1=num_data[num_data['Exited']==1]
num_data_exited_0=num_data[num_data['Exited']==0]
```

```python
num_data_exited_1['Age'].value_counts().sort_index(ascending=True).plot(kind='line')
```

```python
num_data_exited_0['Age'].value_counts().sort_index(ascending=True).plot(kind='line')
```

```python
#compare pattern use common_norm=True (inividually area under curve is 1)
```

```python
sns.kdeplot(data=num_data, x='Age', hue='Exited', palette=['red','blue'], fill=True, common_norm=True)
```

```python
#common_norm=True  show population proportion 
```
