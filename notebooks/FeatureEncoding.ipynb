# Feature Encoding

This notebook demonstrates different techniques used to convert categorical variables into numerical format for machine learning models.

Machine learning algorithms require numerical input, so categorical features such as `Gender` and `Geography` must be encoded before model training.

Techniques covered:
- Label Encoding
- One-Hot Encoding
- Dummy Encoding
- Target Encoding
- Hash Encoding

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, TargetEncoder
```

## Load Dataset

The customer churn dataset is loaded using Pandas.

This dataset contains customer demographic and banking-related information such as geography, gender, credit score, balance, tenure, and churn status.

```python
df=pd.read_csv('customer_churn.csv')
df.head()
```

## Dataset Overview

The `info()` function helps identify:
- number of records
- column names
- data types
- missing values

This is useful for identifying categorical columns that need encoding.

```python
df.info()
```

## Removing Unnecessary Columns

Columns such as `RowNumber`, `CustomerId`, and `Surname` are removed because they are identifiers and do not add meaningful predictive value for this encoding practice.

Removing such columns helps avoid noise in the machine learning workflow.

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True, axis=1)
df.head()
```

## Label Encoding

Label Encoding converts categorical values into numeric labels.

In this example:
- Female → 0
- Male → 1

Label Encoding is suitable when:
- the categorical variable has only two categories
- the encoded numbers do not create misleading order

Here, `Gender` is binary, so Label Encoding is acceptable.

```python
df['Gender'].value_counts()
```

```python
le=LabelEncoder()
df['Gender_Label']=le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
df['Gender_Label'].value_counts()
```

## One-Hot Encoding

One-Hot Encoding creates separate binary columns for each category.

For example, the `Geography` column is converted into:
- France
- Germany
- Spain

Each row receives `True` or `False` values depending on the customer's geography.

This method is useful for nominal categorical variables where there is no natural order.

## Applying One-Hot Encoding to the Dataset

The `pd.get_dummies()` function automatically converts categorical columns into multiple binary columns.

This allows machine learning models to use categorical information in numerical form.

```python
ohe_Geography=pd.get_dummies(df['Geography'])
ohe_Geography
```

```python
ohe=pd.get_dummies(df)
ohe
```

## Dummy Encoding

Dummy Encoding is similar to One-Hot Encoding, but it drops one category using `drop_first=True`.

This helps avoid redundant columns and reduces multicollinearity.

Example:
If `Gender` has two categories, only one dummy column is needed because the other category can be inferred.

```python
de=pd.get_dummies(df,drop_first=True)
df.head()
```

## Target Encoding

Target Encoding replaces each category with the average value of the target variable for that category.

In this example, `Gender` is encoded based on the average churn rate for each gender category.

This technique can be useful when:
- categorical variables have many unique values
- one-hot encoding would create too many columns

Important:
Target Encoding should be used carefully because it can cause data leakage if applied before train-test split.

```python
te=TargetEncoder()
df['GenderEncoder']=te.fit_transform(df[['Gender']],df['Exited'])
df.head()
```

## Hash Encoding

Hash Encoding converts categorical variables into a fixed number of numerical columns using a hashing function.

This is useful when:
- a categorical column has many unique values
- one-hot encoding would create too many columns
- memory efficiency is important

In this notebook, `Geography` and `Gender` are encoded into 5 hashed columns.

# Conclusion

In this notebook, multiple feature encoding techniques were demonstrated.

Key learnings:
- Machine learning models require numerical input.
- Label Encoding is useful for binary or ordinal categories.
- One-Hot Encoding is useful for nominal categorical variables.
- Dummy Encoding reduces redundant columns.
- Target Encoding uses target-based category information but must be used carefully to avoid data leakage.
- Hash Encoding is useful for high-cardinality categorical variables.

Choosing the right encoding technique depends on:
- number of categories
- type of categorical variable
- model type
- risk of multicollinearity
- risk of data leakage
